### MinMax vs Standard vs Robust Scalers

---------------

### Import

```python
import numpy as np
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Expontial distribution

```python
data = np.array([np.random.exponential(10, 10000),
                 np.random.normal(10, 10, 10000),
                 np.random.normal(300, 3, 10000)])
data.shape
```

```python
for i in range(len(data)):
    sns.kdeplot(data[i])
```

### MinMaxScaler

```python
minimum = data.min(axis=1)
maximum = data.max(axis=1)
print(f'Min: {minimum} \nMax: {maximum}')
#Min Max
#Pciture
#RGB
#Red 0-255
```

```python
data_minmax = (data - minimum.reshape(-1,1)) \
            / (maximum - minimum).reshape(-1,1)
```

```python
for i in range(len(data)):
    sns.kdeplot(data_minmax[i])
```

### StandardScaler

```python
means = data.mean(axis=1)
stds = data.std(axis=1)
print(f'Means: {means} \nStandard deviations: {stds}')
```

```python
data_standard = (data - means.reshape(-1,1)) / stds.reshape(-1,1)
```

```python
for i in range(len(data)):
    sns.kdeplot(data_standard[i])
```

### RobustScaler

```python
medians = np.median(data, axis=1)
p75, p25 = np.percentile(data, [75 ,25], axis=1)
iqr = p75 - p25
print(f'Medians: {means} \nIQRs: {iqr}')
```

```python
data_robust = (data - medians.reshape(-1,1)) / iqr.reshape(-1,1)
```

```python
for i in range(len(data)):
    sns.kdeplot(data_robust[i])
```
